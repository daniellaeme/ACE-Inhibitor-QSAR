```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Set up call to CHEMBL activity API
activities = new_client.activity.filter(
    target_chembl_id='CHEMBL1808',          # CHEMBL ID of Human Angiotensin-Converting Enzyme (Single Protein)
    assay_type='B',                         # Filters for Binding Assay (900 to 764)
    standard_type='IC50'                    # Filters for IC50 Activity Data
    ).only([
        'molecule_chembl_id',               # Selects molecule CHEMBL ID
        'target_chembl_id',                 # Selects target CHEMBL ID
        'canonical_smiles',                 # Selects structure
        'standard_type',                    # Selects test type
        'standard_value',                   # Selects raw number
        'standard_units',                   # Selects units
        'pchembl_value',                    # Selects cleaned number
        ])

# Convert list of results to Pandas dataframe
df = pd.DataFrame.from_dict(activities)
print(f"\n{df.shape[0]} rows and {df.shape[1]} columns downloaded.\n")
df.head()
```

```python
df.to_csv('ace_raw.csv', index=False)
print("Data saved to ace_raw.csv successfully." +
      "\n...Data Extraction Complete...")
```
